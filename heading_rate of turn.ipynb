```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
data=pd.read_csv(r"C:\Users\BrunoNad\Downloads\Dataset Fuel Oil(Tabelle1).csv",sep=";")
data
```

```python
headings=data.loc[:,'Heading [degrees]'].dropna().reset_index().drop('index',axis=1)
headings
```

```python
# type object -> type float 
for col in headings.columns:
    if headings[col].dtype == 'object':
        headings[col] = pd.to_numeric(headings[col].str.replace(',', '.'), errors='coerce')
```

```python
headings
```

```python
data_lat_long=data.iloc[:,[1,2]].dropna().reset_index().drop('index',axis=1)
data_lat_long
```

```python
for col in data_lat_long.columns:
    if data_lat_long[col].dtype == 'object':
        data_lat_long[col] = pd.to_numeric(data_lat_long[col].str.replace(',', '.'), errors='coerce')
```

```python
data_lat_long
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data_lat_long.loc[:10,'longitude'], data_lat_long.loc[:10,'latitude'], color='blue')
plt.plot(data_lat_long.loc[:10,'longitude'], data_lat_long.loc[:10,'latitude'], color='red',linewidth=0.6)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```
